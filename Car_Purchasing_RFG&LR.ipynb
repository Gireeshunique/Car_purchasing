```python
# define X  and Y
y = df[['car purchase amount']]
x = df.drop(['car purchase amount'],axis=1)
```

```python
print(x.shape,y.shape)
```

```python
# data pre-processing
from sklearn.preprocessing import MinMaxScaler
mns = MinMaxScaler()
x_scaled = mns.fit_transform(x)
y_scaled = mns.fit_transform(y.values.reshape(-1,1))
```

```python
print(x_scaled.shape,y_scaled.shape)
```

```python
# division data  in training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x_scaled,y_scaled, test_size= 0.2, random_state=41)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
# model and algorithem
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,Y_train)
```

```python
# ANN MODEL
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ml = Sequential()
ml.add(Dense(25, input_dim=9, activation= 'relu'))
ml.add(Dense(25, activation = 'relu'))
ml.add(Dense(1, activation='linear'))
ml.summary()
```

```python
epochs_hist.history.keys()
```

```python
plt.figure(figsize= (10,10), dpi=200)
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss During Training Or Validation')
plt.xlabel('Epochs')
plt.ylabel('Training & Validation Loss')
plt.legend(['Training ', 'Validation Losses'])
```

```python
# alogrithms
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

# Metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# function 

def performance(model, X_train,Y_train,y_pred,Y_test):
     print('Train Score : ', model.score(X_train,Y_train))
     print('Test Score : ', r2_score(Y_test, y_pred))
     print('')
     print('MSE :', mean_squared_error(Y_test,y_pred))
     print('MAE :', mean_absolute_error(Y_test,y_pred))
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)

# prediction data
lr_pred = lr.predict(X_test)
performance(lr,X_train,Y_train, lr_pred,Y_test)
```

```python
# comparision & visulization
plt.scatter(lr_pred,Y_test, c='blue', marker ='o', s=25)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], c ='black', lw =1.5)
plt.xlabel('Prediction Data')
plt.ylabel('Acutal Data')
plt.title('Predict Data VS Actual Data')
plt.show()
```

```python
# fitting model Ridge
ridge = Ridge(alpha = 1)
ridge.fit(X_train, Y_train)

# the predicted data
ridg_pred = ridge.predict(X_test)
performance(ridge,X_train,Y_train, ridg_pred, Y_test)
```

```python
# comparision &visulization Ridge model
plt.scatter(ridg_pred, Y_test, c= 'blue',marker='o', s=25)
plt.plot([Y_test.min(),Y_test.max()],[Y_test.min(),Y_test.max()], c= 'red', lw =2)
plt.xlabel('Prediction Data', c= 'red')
plt.ylabel('Actual Data')
plt.title('Predict Data VS Actual Data',c = 'red')
plt.show()
```
